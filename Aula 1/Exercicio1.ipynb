### Age

Representa a idade do participante

```python
ages = dataset['Age']

ages.value_counts().sort_index().plot(kind='bar', color='lightblue', edgecolor='black')

plt.title("Distribuição de Idades")
plt.xlabel("Idade")

for index, value in enumerate(ages.value_counts().sort_index()):
    plt.text(index - 0.1, value + 2, value)

plt.xticks(rotation=0)
plt.grid(axis='y', linestyle='--')
plt.show()
```

## Relações

Para verificar as relações entre as variáveis foi utilizada a [Phi_K Correlation Constant](https://phik.readthedocs.io/en/latest/#phi-k-correlation-constant)

```python
# numerical features
features_num = []

# categorical features
features_cat = ['Age', 'Gender', 'Occupation', 'Days_Indoors', 'Growing_Stress',
       'Quarantine_Frustrations', 'Changes_Habits', 'Mental_Health_History',
       'Weight_Change', 'Mood_Swings', 'Coping_Struggles', 'Work_Interest',
       'Social_Weakness']

# all features combined
features = features_num + features_cat
```

```python
import phik
import seaborn as sns
```

```python
# calc Phi_K matrix
phiK_mat = dataset[features].phik_matrix(interval_cols=features_num)
```

```python
# visualize phi_K matrix
plt.figure(figsize=(7,5))
sns.heatmap(phiK_mat, annot=True,
            fmt='.2f',
            linecolor='black', linewidths=.5,
            cmap='RdYlGn', vmin=-1, vmax=+1)
plt.title('Phi_K correlation')
plt.show()
```

Com base na matriz de correlação Phi_k é possível perceber que não há muita correlação entre as variáveis. Uma conclusão esperada, visto que, boa parte dos dados da pesquisa estão distribuídos quase que igualmente em suas opções, ou seja, para as perguntas onde as respostas eram Yes/No, as respostas estão sempre próximas de 50/50 e para as perguntas com opções Yes/Maybe/No, as respostas estão próximas de 33/33/33. 

Por causa dessa distribuição estranhamente proporcional dos dados, não é possível ver uma correlação clara entre as variáveis, o que também impede o aproveitamento dos dados para se obter qualquer conclusão relevante sobre os mesmos.
